# Suspicious Kingfisher Art & Glassworks

<font size="2">© Suspicious Kingfisher Art & Glassworks<br><br></font>

Suspicious Kingfisher Art & Glassworks is owned and operated by Cassie Stevenson.


Cassie is a wildlife ecologist and artist, based in Edmonton, Alberta, whose primary medium is stained glass; she also dabbles in drawing, beading, and pottery. Cassie draws inspiration from the natural world and her Metis heritage. Her conservation and ecological background influences many of her pieces, and she aims to foster a deeper appreciation for wildlife and the natural world, and a better balance with it.



*<font size="2">Suspicious Kingfisher Art & Glassworks respectfully acknowledges that we are located on the traditional lands of Treaty 6. This territory has been traditional and ancestral land of the Cree, Dene, Blackfoot, Ojibway/Saulteaux, Nakota Sioux, Métis, and many others whose histories, languages, and cultures continue to influence our vibrant community.</font>*
